- Docstrings are important in the coding process.
- Good documentation of the codes would make the codes earsier to be maintained later on.
- The following example is from [Example NumPy Style Python Docstrings](https://sphinxcontrib-napoleon.readthedocs.io/en/latest/example_numpy.html#example-numpy)
